```python
#Author: Ripa
#penguin species and counts
# Data frames.
import pandas as pd

# Plotting
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import seaborn as sns

#Numerical arrays
import numpy as np

# Warning package to ignore warning messages
import warnings
warnings.filterwarnings("ignore")
# Loading the penguins data set.
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
# We use .null to find the missing values
print(df.isnull().sum())
# Dropping null values.
df.dropna(inplace = True)

# Let's check for null values again
print(df.isnull().sum())
```

```python
#Bar plot for the number of penguins per species.

color= ['blue','green','violet']
df['species'].value_counts().plot(kind='bar',color=color)
plt.title('Penguin species & counts',color="red")


#Add labels
counts=df['species'].value_counts()
for i, count in enumerate(counts):
    plt.text(i,count+1, str(count),ha='center',va='bottom')
```

```python
#barplot for number of penguins in each island



color = ['lightblue', 'purple', 'black']
df['island'].value_counts().plot(kind='bar', color = color)
plt.title('Island and Count', color = "red")

# Add labels
counts = df['island'].value_counts()
for i, count in enumerate(counts):
    plt.text(i, count +1, str(count), ha= 'center', va = 'bottom')

```

```python
#Bar plot showing the classification of species by their gender and bill length.
    
sns.barplot(x="species",y="bill_length_mm",data=df,hue="sex")
plt.title("Bill Length By Species", size=10, color="red")
plt.legend()
```

```python
#bill depth histogram
sns.histplot(x="bill_depth_mm",data=df,hue="species",multiple="stack")
plt.title("Bill depth",size=10,color="red")             
        
```

```python
#flipper length histogram
sns.histplot(x="flipper_length_mm",data=df,hue="species",multiple="dodge")
plt.title("Flipper length",size="10",loc="center",color="red")
```

```python
#flipper length vs body mass correlation
print(df)
sns.scatterplot(x="body_mass_g",y="flipper_length_mm",data=df,hue=("species"),style=("sex"))
plt.title("Flipper length vs body mass",color="red")
```

```python
#coorelation coefficient to assess the relationship between variables
#we must drop first non numerical values

df=df.drop(["species"],axis=1)
df=df.drop(["island"],axis=1)
df=df.drop(["sex"],axis=1)

# to make it easy to read the relationship we round the values to two digits
df.corr().round(2)
```

```python
#let's plot the correlation matrix map

matrix=df.corr().round(2)
sns.heatmap(matrix,annot=True,vmax=1,vmin=-1,center=0, cmap='vlag')

```

```python
#A negative coefficient tells us that the relationship between the variables is negative, meaning that when one value increases, the other decreases.

#Similarly, a positive coefficient indicates that when one value increases, so does the other.

#Bill Length vs. Bill Depth: Weak negative correlation (r = -0.23)
#This implies that as bill length increases, there is a slight tendency for bill depth to decrease, although the correlation isn’t particularly strong.
#Bill Length vs. Flipper Length: A robust positive correlation (r = 0.6562).
#This indicates that as bill length increases, there is a significant tendency for flipper length to also increase.
#Bill Depth vs. Flipper Length: A Moderate negative correlation (r = -0.5839).
#This suggests that as bill depth increases, there is a moderate tendency for flipper length to decrease.
```
